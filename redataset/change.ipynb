```python
import os
import nibabel as nib
import numpy as np
import SimpleITK as sitk
from glob import glob
```

```python
def resize_image_itk(itkimage, newSize, resamplemethod=sitk.sitkNearestNeighbor):

    resampler = sitk.ResampleImageFilter()
    originSize = itkimage.GetSize()  # 原来的体素块尺寸
    originSpacing = itkimage.GetSpacing()
    print(originSpacing)
    newSize = np.array(newSize,float)
    factor = originSize / newSize
    newSpacing = originSpacing * factor
    newSize = newSize.astype(np.int) #spacing肯定不能是整数
    resampler.SetReferenceImage(itkimage)  # 需要重新采样的目标图像
    resampler.SetSize(newSize.tolist())
    resampler.SetOutputSpacing(newSpacing.tolist())
    resampler.SetTransform(sitk.Transform(3, sitk.sitkIdentity))
    resampler.SetInterpolator(resamplemethod)
    itkimgResampled = resampler.Execute(itkimage)  # 得到重新采样后的图像
#     print(itkimgResampled.GetSpacing)
    return itkimgResampled
```

```python
image_path = '..//database//test3/'
image_file = glob(image_path + '*.gz')

for i in range(len(image_file)):
    itkimage = sitk.ReadImage(image_file[i])
#     print(i)
    itkimgResampled = resize_image_itk(itkimage, (512,512,90),resamplemethod= sitk.sitkLinear)#,
    space =itkimgResampled.GetSpacing()
#     print(space)
#     itkimgResampled = resize_image_itk(itkimage, (512,512,90),resamplemethod= sitk.sitkNearestNeighbor) #这里要注意：mask用最近邻插值，CT图像用线性插值
#     print('..//database//retest//data//' + image_file[i][len(image_path):])
    sitk.WriteImage(itkimgResampled,'..//database//retest3/' + image_file[i][len(image_path):])
```
